```python
##workin with large amounts of data like 20gb you can load 100mb once

for df in pd.read_csv('pokemon_data.txt', delimiter = '\t', chunksize=5): # 5 rows down loaded once it can be 100000
    print('chunk df')
    print(df)
```

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('pokemon_data.txt', delimiter = '\t', chunksize=5): # 5 rows down loaded once it can be 100000
    result =df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, result])
    
    print('chunk df')
#     print(result)
    print(new_df)
```
